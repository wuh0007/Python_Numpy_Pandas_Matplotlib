```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1,7):#draw 6 lines
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
```

```python
sns.set()
sinplot()
```

Five Themes Styles
1. darkgrid
2. whitegrid
3. dark
4. white
5. ticks

```python
sns.set_style("whitegrid")
data = np.random.normal(size=(20,6)) + np.arange(6)/2
sns.boxplot(data=data)
sns.despine()
```

```python
sns.set_style("dark")
sinplot()
```

```python
sns.set_style("white")
sinplot()
```

```python
sns.set_style("ticks")
sinplot()
```

```python
sinplot()
sns.despine()
```

```python
sns.violinplot(data)
sns.despine(offset=30)
```

```python
sns.set_style("whitegrid")
sns.boxplot(data=data, palette="deep")
# sns.despine(left=True)
```

```python
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

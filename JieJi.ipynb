```python
import pandas as pd
import numpy as np
from scipy import stats
import math 

def geometricMean(arr, n): 
      
    # declare sum variable and 
    # initialize it to 1. 
    sum = 0; 
      
    # Compute the sum of all  
    # the elements in the array. 
    for i in range(n): 
        sum = sum + math.log(arr[i]+1); 
      
    # compute geometric mean  
    # through formula antilog 
    # (((log(1) + log(2) + . . 
    # ... + log(n))/n) 
    # and return the value to  
    # main function. 
    sum = sum / n; 
      
    return math.exp(sum)-1; 
info = pd.read_excel("Pravda Asset Management data - updated.xlsx")
info.head()
```

```python
selected_columns = info[['3-Mo. T-Bill\nReturn\n(Annualized)', 'Real S&P\nComposite\nAnnual\nReturn','Real\nSmoothed\nP/Avg10(E)']]
# print(selected_columns)
greater_than_12_rows = selected_columns.loc[selected_columns['Real\nSmoothed\nP/Avg10(E)'] > 12]
a= greater_than_12_rows[['Real S&P\nComposite\nAnnual\nReturn']]
a.columns = range(a.shape[1])
less_than_12_rows = selected_columns.loc[selected_columns['Real\nSmoothed\nP/Avg10(E)'] <= 12]
b = less_than_12_rows[['3-Mo. T-Bill\nReturn\n(Annualized)']]
b.columns = range(b.shape[1])
c = a.append(b)
print(c)
c = c.values
n = len(c) 
print(geometricMean(c, n)); 
```
